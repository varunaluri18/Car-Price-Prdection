### Feature Scailing<a href="#Feature-Scailing" class="anchor-link">¶</a>

-   The algorithms having gradient discent(LinearRegression),Eucledian
    distance(knn) requires feature scailing. The data points should be
    in the graph limitation range so we scale the features magnitude to
    be fitted in the graph.

-   Ensemble techniques(decision forest, random forest, xg boost) dont
    require feature scailing.

***Standardization vs Normalization***

The main purpose of this scales are to rescale the values of features in
the same unit so that in the obtained distribution every point or value
is close to each other.

-   standardization and normalization are scailing techniques which are
    applied on an algorithm to make features of data into our required
    scale(i.e units and magnitude). So, that features become easy to
    apply on the model.

-   Units and magnitude (ex:25yrs --\> yrs here is units and 25 is our
    magnitude)

<!-- -->

-   Standardization: it helps to scale down the features of data based
    on standard and normal distribution curve with mean=0 & standard
    deviation=1

    -   range: bell curve
    -   Xstand = (X-mean) / (sd)
    -   StandardScaler() also called as Z-Score Normalization

<!-- -->

-   Normalization: it helps to scale down features of data between 0 to
    1

    -   range : 0 to 1
    -   Xnorm = (X-Xmin) / (Xmax-Xmin)
    -   MinMaxScaler()

-   Reference Link :
    <https://www.analyticsvidhya.com/blog/2020/04/feature-scaling-machine-learning-normalization-standardization/>
